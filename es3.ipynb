import folium

coordinate = []

for m in meteoriti:
  lat = m.get("lat")
  lon = m.get("long")

coordinate.append(lat, lon, m)

media_lat = sum(c[0] for c in coordinate) / len(coordinate)
media_lon = sum(c[1] for c in coordinate) / len(coordinate)

mappa = folium.Map(location=[media_lat, media_lon], zoom_start=2, tiles="OpenStreetMap")

for lat, lon, m in meteoriti:
    nome = m.get("name")

    folium.Marker(
        location=[lat, lon],
        popup=nome
    ).add_to(mappa)

file = "mappa_meteoriti.html"
mappa.save(file)
