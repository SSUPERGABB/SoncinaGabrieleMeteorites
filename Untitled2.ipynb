```python
import matplotlib.pyplot as plt

masse = []

for m in meteoriti:
  valore = m["mass"]
  masse.append(valore)

plt.figure()
plt.hist(masse, bins= 30, color = "purple", edgecolor = 'black')
plt.title("Distribuzione delle masse dei meteoriti")
plt.xlabel("Massa (grammi)")
plt.ylabel("Meteoriti")
plt.show()
```
